```python
#3
import pandas as pd

# Load your dataset
df = pd.read_csv("https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-05-05/villagers.csv")

# Get basic information about the DataFrame
print("DataFrame Information:")
print(df.info())

# Get summary statistics for numerical and categorical columns
print("\nSummary Statistics:")
print(df.describe(include='all'))

# Get data types of each column
print("\nColumn Data Types:")
print(df.dtypes)

# Get count of missing values per column
print("\nMissing Values Count:")
print(df.isna().sum())

# Example of getting value counts for a specific categorical column
print("\nValue Counts for 'gender':")
print(df['gender'].value_counts())
```
